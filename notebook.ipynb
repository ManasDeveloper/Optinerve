```python
import ast

def extract_single_alphabet(label_str):
    try:
        # Convert string to list
        label_list = ast.literal_eval(label_str)

        # Return first alphabet character (if any)
        for item in label_list:
            for char in item:
                if char.isalpha():
                    return char
    except:
        pass
    return None  # or return '' if you prefer
```

```python
data_new['labels'] = data_new['labels'].apply(extract_single_alphabet)
```
